```python
import json
from openaq import OpenAQ
```

```python
with open("../secret.json") as f:
    secret = json.load(f)
    f.close()

openaq_key= secret["openaq-api-key"]
```

```python
client=OpenAQ(api_key=openaq_key)
client
```

```python
ind_bbox_search_locations=client.locations.list(
    bbox=(75.405762,27.118238,78.855469,29.219120),
    limit=1000
)
ind_bbox_search_locations.meta
```

```python
ind_bbox_search_locations.results
```

```python
ind_bbox_search_locations.results[0].__dict__
```

```python
location_info={}
for location in ind_bbox_search_locations.results:
    location_info[location.id] = location.name

location_info
```

```python
len(location_info)
```

```python
with open("../locations.json", "w", encoding="utf-8") as f:
    json.dump(location_info, f, ensure_ascii=False, indent=4)
    f.close()
```
